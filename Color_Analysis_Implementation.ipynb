# An example on image color analysis

```python
from color_analysis import multiple_images, rgb_histogram, rgb_image, colors_plot, elbow_curve, color_clustering, save_clusters_info, ImageAnalizer
```

### The image

```python
path = "Imgs/1bd9b38e59eabb23abbcda38a323c8e9.jpg"
Imagen = ImageAnalizer(path)
Imagen.img
```

### RGB components

```python
Imagen.RGB_graph()
```

### Generating the color sample with kmeans

```python
maxclusters = 30
clusters = elbow_curve(maxclusters, Imagen.pixel_values)
```

```python
Imagen.Color_palette(clusters)
```
