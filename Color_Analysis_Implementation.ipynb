# An example on image color analysis

```python
from color_analysis import multiple_images, elbow_curve, color_clustering, ImageAnalizer
```

### The image

```python
path = "Imgs/fe7e1edbcda543fc2d7e3dc4bf54c5fe.jpg"
Imagen = ImageAnalizer(path)
Imagen.img
```

### RGB components

```python
Imagen.rgb_histogram()
```

```python
Imagen.rgb_composite()
```

### Generating the color sample with kmeans

```python
maxclusters = 30
clusters = elbow_curve(maxclusters, Imagen.pixel_values)
```

```python
# parameters: the number of clusters, and the output folder
Imagen.color_palette(clusters, "ColorPalette_withKnee")
```
