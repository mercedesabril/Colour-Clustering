# An example on image color analysis

```python
from color_analysis import *
```

### The image

```python
path = "Imgs/jean.jpg"
Imagen = ImageAnalizer(path)
Imagen.img
```

### RGB components

```python
Imagen.RGB_graph()
```

### Generating the color sample with kmeans

```python
maxclusters = 30
clusters = ElbowCurve(maxclusters, Imagen.pixel_values)
```

```python
Imagen.Color_palette(clusters)
```
