# Sorting images by their principal colors

```python
import pandas as pd
from PIL import Image
import math
import colorsys
import os
```

## Load the datasets

```python
df_col = pd.read_csv("ColorClusterRGB_with_knee.csv")
df_per = pd.read_csv("ColorClusterPercentage_withKnee.csv")
```

## Extract the maximum value per column

```python
first_max = []
second_max = []
imgs_name = []
for i in df_col.keys():
    imgs_name = imgs_name + [i]
    maxs = df_per[i].nlargest(2).keys()
    first_max = first_max + [df_col[i][maxs[0]]]
    second_max = second_max + [df_col[i][maxs[1]]]
```

```python
from ast import literal_eval as make_tuple

first_max = list(map(make_tuple, first_max))
second_max = list(map(make_tuple, second_max))
```

## Color strip image

```python
def color_stripe(size, values, name):
    '''
    Color Sampler
    
    - Description 
        Saves a stripe of RGB colors of desired length.
    
    - Parameters
        size: a list of width and height values
        values: the RGB values to plot
    
    '''
    
    if not os.path.isdir("color_stripe"):
        os.makedirs("color_stripe")
        
    sample = []
    width, height = size
    temp = Image.new('RGB', (width, height))
    new_img = Image.new('RGB', (width*len(values), height))
    x_offset = 0 
    
    for val in values:
        sample = [val] * ((width*height))
        temp.putdata(sample)
        new_img.paste(temp, (x_offset,0))
        x_offset = x_offset + width
    
    new_img.save("color_stripe/"+ name + '.jpg')
```

## Color sorting

```python
class SortedColors(object):
    
    def __init__(self, DATA, sortby):
        self.dataset = list(zip(*DATA))
        self.sorted = sorted(self.dataset, key=lambda row: row[sortby])
        self.sorted_RGBvalues = [x[1] for x in self.sorted]
        self.sorted_names = [x[0] for x in self.sorted]
        
```

### RGB sorting

The most simple way of sorting. First, it sorts the first component (R: red), then the second (G: green) and finally the third (B: blue) of the principal color of the image.

```python
RGB = SortedColors((imgs_name, first_max), 1)
color_stripe([20,100], RGB.sorted_RGBvalues, 'ColorStripe_RGB')
RGB.sorted[0:10]
```

### HSV sorting

The HSV color space represents the similarity between colors better than the RGB space. It uses hue, saturation and value as parameters.

```python
def hsv(x): 
    return(colorsys.rgb_to_hsv(*x))
```

```python
hsv_t = [hsv(i) for i in first_max]
HSV = SortedColors((imgs_name, first_max, hsv_t), 2)
color_stripe([20,100], HSV.sorted_RGBvalues, 'ColorStripe_HSV')
HSV.sorted[0:10]
```

### HLS sorting

HLS is similar to HSV, but instead the color space is organized by hue, saturation and lightness

```python
def hls(x):
    return(colorsys.rgb_to_hls(*x))
```

```python
hls_t = [hls(i) for i in first_max]
HLS = SortedColors((imgs_name, first_max, hls_t), 2)
color_stripe([20,100], HLS.sorted_RGBvalues, 'ColorStripe_HLS')
HLS.sorted[0:10]
```

### Luminosity sorting

Sorts by the percieved luminosity of the color

```python
def lum (r,g,b):
    return math.sqrt( .241 * r + .691 * g + .068 * b )
```

```python
lum_t = [lum(*i) for i in first_max]
LUM = SortedColors((imgs_name, first_max, lum_t), 2)
color_stripe([20,100], LUM.sorted_RGBvalues, 'ColorStripe_LUM')
LUM.sorted[0:10]
```

### Step sorting

Based on the HSV color space, but also sorts by luminosity, generating a number of color segments. The ideal number of segments varies from dataset to dataset. To add continuity to the segments, the second step method inverts the luminosity every two segments.

```python
def step (r,g,b, repetitions=1):
    
    lum = math.sqrt( .241 * r + .691 * g + .068 * b )
    h, s, v = colorsys.rgb_to_hsv(r,g,b)
    
    h2 = int(h * repetitions)
    lum2 = int(lum * repetitions)
    v2 = int(v * repetitions)
    
    return (h2, lum, v2)
```

```python
for number_steps in range(0,20):
    step_t = [step(*i, number_steps) for i in first_max]
    STEP = SortedColors((imgs_name, first_max, step_t), 2)
    color_stripe([20,100], STEP.sorted_RGBvalues, 'ColorStripe_STEP' + str(number_steps))
```

```python
def step_lum (r,g,b, repetitions=1):
    
    lum = math.sqrt( .241 * r + .691 * g + .068 * b )
    h, s, v = colorsys.rgb_to_hsv(r,g,b)
    
    h2 = int(h * repetitions)
    lum2 = int(lum * repetitions)
    v2 = int(v * repetitions)
    
    if h2 % 2 == 1:
        v2 = repetitions - v2
        lum = repetitions - lum
        
    return (h2, lum, v2)
```

```python
for number_steps in range(0,20):
    steplum_t = [step_lum(*i, number_steps) for i in first_max]
    STEP_LUM = SortedColors((imgs_name, first_max, steplum_t), 2)
    color_stripe([20,100], STEP_LUM.sorted_RGBvalues, 'ColorStripe_STEPLUM' + str(number_steps))
```

## Plot the sorted images

```python
from PIL import ImageOps
```

```python
final_t = [step_lum(*i, 6) for i in first_max]
FINAL = SortedColors((imgs_name, first_max, final_t), 2)
```

```python
def collage(images, size, rows, folder):

    final_img = Image.new('RGB', (math.ceil(len(images)/rows)*size, rows*size))
                
    x_offset = 0
    y_offset = 0
    count = 0
    
    for file in images: 
        
        img = Image.open(folder + "/" + file + ".jpg")
        thumb = ImageOps.fit(img, (size, size), Image.ANTIALIAS)
        
        final_img.paste(thumb, (x_offset,y_offset))
        
        y_offset = y_offset + size
        count = count + 1
        
        if count % rows == 0:
            x_offset = x_offset + size
            y_offset = 0
            

    final_img.save("Collage_withKnee.jpg")
    
```

```python
collage(FINAL.sorted_names, 400, 4, "Imgs_done")
```
